```python
import json

purchases = {}

with open('purchase_log.txt', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        data = json.loads(line)
        user_id = data['user_id']
        category = data['category']
        purchases[user_id] = category

with open('visit_log.csv', encoding='utf-8') as visits, \
     open('funnel.csv', 'w', encoding='utf-8') as funnel:

    # заголовок
    funnel.write('user_id;source;category\n')

    for line in visits:
        user_id, source = line.strip().split(';')
        
        if user_id in purchases:
            category = purchases[user_id]
            funnel.write(f'{user_id};{source};{category}\n')
```
