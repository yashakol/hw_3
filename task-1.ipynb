```python
import json

purchases = {}

with open('purchase_log.txt', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        data = json.loads(line)          
        user_id = data['user_id']        
        category = data['category']      
        purchases[user_id] = category    

i = 0
for k, v in purchases.items():
    print(f"{k} '{v}'")
    i += 1

```
